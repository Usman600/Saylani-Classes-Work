```python


from bs4 import BeautifulSoup
import requests

url = "https://realpython.github.io/fake-jobs/"
request = requests.get(url)
request





    
```

```python
info = BeautifulSoup(request.content,"html5lib")
info
```

```python
title = info.find(class_="title is-1")
title.text.strip()
```

```python
div = info.find(id="ResultsContainer")
div.text.strip()
```

```python
jobs = info.find_all(class_="column is-half")
jobs

for job in jobs:
    print(job.text.strip())
```

```python
job_title = info.find_all(class_="title is-5")
job_title

for title in job_title:
    print(title.text.strip())
```

```python
subtitle = info.find_all(class_="subtitle is-6 company")
subtitle

for sub in subtitle:
    print(sub.text.strip())
```

```python
Location = info.find_all(class_="location")
Location
```

```python
date = info.find_all(datetime="2021-04-08")
date
```
